# Indexing

Two primary operators: **loc** (label-based) and **iloc** (index-based selection).


```python
# reviews.iloc[9] #Presents row then column
reviews.iloc[:3,:2] # [entries, properties]
# Python-like syntax. [1,2,3], :10, -5: all valid to get specific entries
```

```python
# loc to index by value not position
reviews.loc[0, 'country']
```

Note: iloc follows python index convention; loc indexes inclusively (i.e. will include final value)

**map()** to (as one would imagine) map one set of values to another.
* Your mapping functionj should go from single value -> single value. map() returns a new, transformed, Series.
* **apply()** is the equivalent for a whole DataFrame
* Simple functions like < > == etc. are built into Pandas and operate quicker, but map and apply have more expressivity

Both of these methods return new Series/DFs i.e. they don't modify the original

```python
reviews.points.map(lambda p: p - reviews.points.mean())
```

```python
def remean(row):
  row.points = row.points - reviews.points.mean()
  return row

reviews.apply(remean, axis='columns').head() #axis='index'=>function that transforms each column
```

# Groupwise Analysis


```python
reviews.groupby('points').points.count().head()
```

```python
reviews.value_counts('points').head()
```

```python
# Value counts = grouping function
reviews.value_counts('points').min()
```

```python
# Groups the bp by different points values, then looks at the minimum for the desired feature within each group
reviews.groupby('points').price.min()
```

```python
reviews.groupby('winery').apply(lambda p: p.title.iloc[0])
# New index is the first wine to be reviewed for each winery
# Group entries by winery then let the title be the first entry
```

```python
 # You can group by multiple things
 reviews.groupby(['country', 'province']).apply(lambda p: p.loc[p.points.idxmax()])
 #So now we can see the maximum value for each province
```

```python
reviews.groupby(['country']).price.agg([len, min, max])
#agg to run multiple functions in one call
```
